```python
summary = []

for col in df1.columns:
    dtype = df1[col].dtype
    unique = df1[col].nunique(dropna=True)
    missing = df1[col].isna().sum()
    
    if pd.api.types.is_numeric_dtype(df1[col]):
        dist = f"min={df1[col].min()}, max={df1[col].max()}, mean={round(df1[col].mean(),2)}"
        significance = "Measure"
    else:
        top = df1[col].astype(str).str.upper().value_counts().head(3).to_dict()
        dist = f"Top values: {top}"
        significance = "Attribute"
    
    summary.append([col, str(dtype), unique, missing, dist, significance])

summary_df = pd.DataFrame(summary, columns=[
    "Column",
    "Data Type",
    "Unique Values",
    "Missing Count",
    "Distribution Snapshot",
    "Attribute/Measure"
])

summary_df.to_excel('column_summary.xlsx')
```

```python
#we change  textual categorical data description columns to lowercase to keep uniformity
categorical_cols = ['CORRECTION_VERBATIM','CUSTOMER_VERBATIM','GLOBAL_LABOR_CODE_DESCRIPTION']
import re
#detect not english characters in these columns
def contains_non_english(text):
    if pd.isna(text):
        return False
    return bool(re.search(r'[^\x00-\x7F]', str(text)))
df1['has_non_english'] = False
df1['has_non_english'] |= df1['CORRECTION_VERBATIM'].apply(contains_non_english)
df1['has_non_english'] |= df1['CUSTOMER_VERBATIM'].apply(contains_non_english)
df1['has_non_english'] |= df1['GLOBAL_LABOR_CODE_DESCRIPTION'].apply(contains_non_english)
non_english_rows = df1[df1['has_non_english']]
print(len(non_english_rows))
print(non_english_rows[['CORRECTION_VERBATIM', 'CUSTOMER_VERBATIM','GLOBAL_LABOR_CODE_DESCRIPTION','has_non_english']].head())
```

```python

from deep_translator import GoogleTranslator

def translate_to_english(text):
    if not isinstance(text, str) or text.strip() == "":
        return text
    
    return GoogleTranslator(
        source="auto",
        target="en"
    ).translate(text)

#translated = translate_to_english("Bonjour")
#print(translated)

for idx, row in df1[df1['has_non_english']==True].iterrows():
    df1.at[idx, 'CORRECTION_VERBATIM'] = translate_to_english(row['CORRECTION_VERBATIM'])
    df1.at[idx, 'CUSTOMER_VERBATIM'] = translate_to_english(row['CUSTOMER_VERBATIM'])
    df1.at[idx, 'GLOBAL_LABOR_CODE_DESCRIPTION'] = translate_to_english(row['GLOBAL_LABOR_CODE_DESCRIPTION'])
    #print(row['CORRECTION_VERBATIM'])
   
```

```python
#now we convert to lowercase
for col in categorical_cols:
    df1[col].replace(np.nan, 'UNKNOWN', inplace=True)
    df1[col] = df1[col].str.lower()
```

```python
df1.drop(columns=['has_non_english'], inplace=True)
```

```python
#cosmetic changes to some categorical columns to retain alphanumeric format
def normalize_text(text):
    if not isinstance(text, str):
        return text
    
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)        # normalize spaces
    text = re.sub(r'[^\w\s]', '', text)     # remove punctuation
    return text.strip()

for col in categorical_cols:
    df1[col] = df1[col].apply(normalize_text)
```

```python
#to retain alpha numeric format 
def keep_alphanumeric(text):
    if not isinstance(text, str):
        return text
    return re.sub(r'[^A-Za-z0-9]', '', text)
df1['REPAIR_DLR_POSTAL_CD'] = df1['REPAIR_DLR_POSTAL_CD'].apply(keep_alphanumeric)
df1['PLATFORM'] = df1['PLATFORM'].apply(keep_alphanumeric)
df1['LINE_SERIES'] = df1['LINE_SERIES'].apply(keep_alphanumeric)
```

GENERATING TAGs

```python
#We generate tags based on CUSTOMER_VERBATIM and CORRECTION_VERBATIM
def generate_tags(row):

    customer_note = str(row['CUSTOMER_VERBATIM']).lower()
    tech_note = str(row['CORRECTION_VERBATIM']).lower()
    text = f"{customer_note} {tech_note}"
    
    # category1: how it failed
    tags = []
    if any(k in text for k in ['peel', 'delaminat', 'bubble', 'wear', 'flaking', 'finish','peeling', 'fading', 'discolor', 'scratch', 'blemish']):
        tags.append('Surface Degradation')
    if any(k in text for k in ['stitch', 'seam', 'thread', 'fraying', 'sewing','woven', 'fabric', 'leather', 'material', 'wrap']):
        tags.append('Stitching/Material Issue')
    if any(k in text for k in ['heat', 'warm', 'cold', 'thermal', 'temp', 'element', 'heater', 'heated', 'lukewarm', 'temperature']):
        tags.append('Thermal/Heating Failure')
    if any(k in text for k in ['grind', 'creak', 'vibration', 'buzz', 'noise', 'click', 'scrap', 'squeak', 'rattle', 'clunk', 'rubbing']):
        tags.append('NVH/Mechanical Noise')
    if any(k in text for k in ['inop', 'not work', 'malfunction', 'short','circuit', 'code', 'dtc', 'error', 'fault',  'diagnostic', 'programming', 'module','fail','failure']):
        tags.append('Electrical/Electronic Failure')
    if any(k in text for k in ['loose', 'gap', 'fitment', 'align', 'deform','coming off', 'coming apart', 'detach', 'separate', 'misalign', 'crooked', 'off center','broke','broken']):
        tags.append('Fitment/Alignment Issue')

    # category2: which component failed
    components = []
    if any(k in text for k in ['wheel', 'rim', 'spoke', 'grip', 'handle']):
        components.append('Steering Wheel Rim')
    if any(k in text for k in ['switch', 'button', 'control', 'paddle', 'dial','knob', 'scroll', 'selector']):
        components.append('Controls/Switches')
    if any(k in text for k in ['horn', 'contact', 'beep', 'honk', 'beep']):
        components.append('Horn System')
    if any(k in text for k in ['super cruise', 'sensor', 'led', 'light bar', 'driver assist', 'cruise', 'adaptive', 'lane','autonomous', 'assist system']):
        components.append('ADAS/SuperCruise')
    if any(k in text for k in ['heater', 'element', 'module', 'wiring','harness', 'connector', 'terminal']):
        components.append('Heating Module/Element')
    if any(k in text for k in ['airbag', 'cover', 'trim', 'bezel', 'applique','panel', 'plastic', 'chrome']):
        components.append('Airbag/Trim/Chrome')
    if any(k in text for k in ['clock spring', 'spiral cable', 'coil', 'rotary']):
        components.append('Clock Spring/Spiral Cable')

    #category3 : severity tags 
    severity = []
    if any(k in text for k in ['urgent', 'immediate', 'asap','airbag', 'detach', 'explosive', 'deploy', 'safety']):
        severity.append('Critical Severity')
    if any(k in text for k in ['inop', 'not work', 'failure', 'broken', 'malfunction']):
        severity.append('Functional Failure')
    if any(k in text for k in ['peel', 'scratch', 'blemish', 'finish', 'appearance', 'fading', 'discolor', 'rattle', 'squeak', 'vibration', 'noise', 'creak', 'grind']):
        severity.append('Cosmetic Issue')
    if any(k in text for k in ['intermittent', 'sometimes', 'occasional', 'sporadic']):
        severity.append('Inttermittent Severity')
    if any(k in text for k in ['always', 'constant', 'persistent', 'continuous', 'minor', 'slight', 'small', 'low', 'negligible', 'insignificant']):
        severity.append('Persistent Issues')
    
    # default
    condition_final = ", ".join(tags) if tags else "General Investigation"
    component_final = ", ".join(components) if components else "General Assembly"
    severity_final = ", ".join(severity) if severity else "Intermittent Severity"

    return pd.Series([condition_final, component_final, severity_final])
```

```python
import seaborn as sns

df1['REPAIR_DATE'] = pd.to_datetime(df1['REPAIR_DATE'])

# 1. GLOBAL_LABOR_CODE_DESCRIPTION(Steering Wheel Replacement) vs BODY_STYLE Pie Chart
target_labor = 'steering wheel replacement'
filtered_df = df1[df1['GLOBAL_LABOR_CODE_DESCRIPTION'].str.lower() == target_labor]
body_style_counts = filtered_df['BODY_STYLE'].value_counts()

plt.figure(figsize=(10, 7))
colors = sns.color_palette('pastel')[0:len(body_style_counts)]
plt.pie(
    body_style_counts, 
    labels=body_style_counts.index, 
    autopct='%1.1f%%', 
    startangle=180, 
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)

plt.title(f'Vehicle Body Styles Affected by\n"{target_labor.title()}"', fontsize=14, pad=20)
plt.axis('equal')  
plt.legend(body_style_counts.index, title="Body Styles", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()
plt.show()
```
